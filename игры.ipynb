***Описание проекта***

Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). **NB**: Данные за 2016 год могут быть неполными.

***Описание данных*** 

* Name — название игры
* Platform — платформа
* Year_of_Release — год выпуска
* Genre — жанр игры
* NA_sales — продажи в Северной Америке (миллионы проданных копий)
* EU_sales — продажи в Европе (миллионы проданных копий)
* JP_sales — продажи в Японии (миллионы проданных копий)

***Рейтинг игр ESRB:***

- «EC» («Early childhood») — «Для детей младшего возраста»;
- «E» («Everyone») — «Для всех»;
- «E10+» («Everyone 10 and older») - «Для всех от 10 лет и старше»;
- «T» («Teen») — «Подросткам»;
- «M» («Mature») — «Для взрослых»;
- «AO» («Adults Only 18+») — «Только для взрослых»;
- «RP» («Rating Pending») — «Рейтинг ожидается»;
- «RP 17+» («Rating Pending») — «Рейтинг ожидается - примерно 17+».

```python
# напишем функцию для визуализации количества пропусков в данных

def pass_value_barh(df):

    fig, ax = plt.subplots(figsize=(19, 6))

    try:
        (
            (df.isna().mean()*100)
            .to_frame()
            .rename(columns = {0:'space'})
            .query('space > 0')
            .sort_values(by='space', ascending=True)
            .plot(kind='barh', figsize=(19,6), legend=False, fontsize=16, ax = ax)

        );

        plt.title('Пропущенные значения, %' + "\n", fontsize=22)
        plt.xlabel('Процент пропусков', fontsize=15)
        plt.ylabel('Столбцы с пропусками', fontsize=15)

        plt.xlim(0, 100)

        for i in ax.patches:
            plt.text(i.get_width()+0.2, i.get_y()+0.1,
                     str(round((i.get_width()), 3)),
                     fontsize=14,
                     color='black')

    except:
        print('пропусков не осталось или произошла ошибка в первой части функции')
```

```python
# выведем барплот с количеством пропусков
pass_value_barh(games)
```

#### Выясним, как менялись продажи по платформам 

Выберем платформы с наибольшими суммарными продажами

```python
# создадим сводную таблицу по платформам и продажам и построим барплот
games.pivot_table(index='platform',
                  values='total_sales', 
                  aggfunc='sum').sort_values(by='total_sales').plot(kind='bar',
                                                                    figsize=(15,5),
                                                                    legend=False) 

plt.grid(axis='y')
plt.title('Суммарное количество продаж игр по платформам'+ "\n", fontsize=20)
plt.xlabel('Платформы', fontsize=15)
plt.ylabel('Количество проданных копий игр, млн', fontsize=15)
plt.show()
```

Выделяются следующие платформы: DS, PS, PS2, PS3, Wii, X360. Сравним продажи по годам для этих платформ:

```python
# создадим датафрейм с интересуемыми нас платформами
platforms = games.query('(platform == ["DS", "PS", "PS2", "PS3", "Wii", "X360"]) & (year_of_release != -1)')

# создадим сводную таблицу с платформами, годами и продажами
platforms = platforms.pivot_table(index=['platform', 'year_of_release'], 
                                  values='total_sales', 
                                  aggfunc='sum')
platforms.reset_index(inplace=True)

# построим графики продаж по годам для каждой платформы
plt.figure(figsize=(15, 5))
plt.title('Сравнение продаж игр топ-6 платформ по годам', fontsize=20)
sns.lineplot(x="year_of_release",
             y="total_sales",
             data=platforms,
             hue="platform")

plt.xlabel('Годы', fontsize=15)
plt.ylabel('Продажи, млн', fontsize=15)
plt.xticks(ticks=range(platforms['year_of_release'].min(), platforms['year_of_release'].max()+1, 2))
plt.grid(True)

plt.show()
```

#### Исследуем, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Построим диаграмму рассеяния и посчитаем корреляцию между отзывами и продажами. 

Для анализа возьмем консоль PS4:

```python
# создадим отдельный датафрейм с данными для PS4
ps4_stat = actual_data.query('(platform=="PS4")&(critic_score !=-1)&(user_score !=-1)')
```

```python
# найдем коэффициент корреляции между оценками пользователей и общими продажами
correlation1 = ps4_stat["user_score"].corr(ps4_stat["total_sales"])

# найдем коэффициент корреляции между оценками пользователей и общими продажами
correlation2 = ps4_stat["critic_score"].corr(ps4_stat["total_sales"])
```

```python
# диаграмма рассеяния для корреляции оценки критиков и общих продаж
ps4_stat.plot(kind='scatter',
              y='total_sales' , 
              x='critic_score', 
              figsize=(10,4))
plt.title('Диаграмма рассеяния: Оценка критиков — Количество продаж', fontsize=15)
plt.xlabel('Оценки критиков')
plt.ylabel('Количество продаж, млн');
plt.show()
print(f'Коэффициент корреляции между оценками критиков и общими продажами: {correlation2:.3f}')
```

```python
# диаграмма рассеяния для корреляции оценки пользователей и общих продаж
ps4_stat.plot(kind='scatter',
              y='total_sales' , 
              x='user_score', 
              figsize=(10,4))
plt.title('Диаграмма рассеяния: Оценка пользователей — Количество продаж',fontsize=15)
plt.xlabel('Оценки критиков')
plt.ylabel('Количество продаж, млн')
plt.show()
print(f'Коэффициент корреляции между оценками пользователей и общими продажами: {correlation1:.3f}')
```

**Вывод:**
Если оценивать по шкале Чеддока, между оценками пользователей и количеством продаж корреляции нет, а между оценками критиков и общими продажами связь прямая умеренная.

#### Выясним, как данный вывод соотносится  с данными для других платформ

```python
# создадим датафрейм с исследуемыми платформами в котором нет "заглушек"
scores_and_sales = selected_platforms.query('(critic_score !=-1)&(user_score !=-1)')

# напишем функцию для построения диаграмм рассеяния и вычисления коэффициентов корреляции
def correlation(df, platform_name):
    
    # создаем датафрейм для платформы
    platform = df[df['platform']==platform_name]
    
    # задаем расположение графиков
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4)) 
    
    # строим диаграмму рассеяния "Оценки пользователей - Количество продаж"
    platform.plot(kind='scatter',
                  y='total_sales', 
                  x='user_score', 
                  ax = ax1)
    ax1.set_title(platform_name)
    ax1.set_xlabel('Оценки пользователей')
    ax1.set_ylabel('Количество продаж, млн')
    
    # строим диаграмму рассеяния "Оценки критиков - Количество продаж"
    platform.plot(kind='scatter',
                  y='total_sales', 
                  x='critic_score', 
                  ax = ax2)
    ax2.set_title(platform_name)
    ax2.set_xlabel('Оценки критиков')
    ax2.set_ylabel('Количество продаж, млн')
    
    plt.tight_layout()  
    plt.show()
    
    # вычисляем коэффициент корреляции
    corr1 = platform["user_score"].corr(platform["total_sales"])
    corr2 = platform["critic_score"].corr(platform["total_sales"])
    
    # выводим значения коэффициента корреляции на экран
    print(f'Коэффициент корреляции между оценками пользователей и общими продажами: {corr1:.3f}')
    print(f'Коэффициент корреляции между оценками критиков и общими продажами: {corr2:.3f}')
```

```python
# применим функцию для исследуемых  платформ
for platform in ["XOne","3DS","PC","WiiU","PSV"]:
    correlation (scores_and_sales, platform)
```

***Вывод:***

- Связь между оценками пользователей и общими продажами в большинстве случаев отсутствует, для 3DS данная связь слабая, а для WiiU - умеренная.

- Связь между оценками критиков и общими продажами у XOne, 3DS и WiiU умеренная, в остальных случаях она слабая или отсутствует совсем.

- В целом между оценками и продажами нет четко выраженной устойчивой связи. В большей степени она проявляется между оценками критиков и продажами.

Все верно, корреляция между продажи и отзывами пользователей отсутствует и очень слабая с отзывами критиков. Стоит присматриваться к экспертному мнению.

###  Составим портрет пользователя каждого региона

***Определим для пользователя каждого региона (NA, EU, JP):***
* Самые популярные платформы (топ-5). 
* Самые популярные жанры (топ-5).
* Влияет ли рейтинг ESRB на продажи в отдельном регионе?

Напишем функцию для определения характеристик каждого региона:

```python
def user_portrait(df, sales_column, region):
    
    # сводная таблица для поиска самых популярных платформ в регионе
    df_platform = df.pivot_table(index='platform', 
                                 values=sales_column, 
                                 aggfunc='sum').sort_values(by=sales_column, 
                                                            ascending=True)
    
    # барплот для поиска самых популярных платформ в регионе
    ax = df_platform.plot(kind='barh', legend=False, figsize=(19, 8), color ='blue')

    for i in ax.patches:
        ax.text(i.get_width() + 0.2, i.get_y() + i.get_height() / 2,
                str(round((i.get_width()), 3)),
                fontsize=14,
                color='black')
    plt.title(f'{region}: Количество продаж игр по платформам за 2012-2016 гг.'+ "\n", fontsize=20)
    plt.xlabel('Количество проданных копий игр, млн', fontsize=15)
    plt.ylabel('Платформы', fontsize=15)
    plt.show()

    # сводная таблица для поиска самых популярных жанров в регионе
    df_genre = df.pivot_table(index='genre', 
                              values=sales_column, 
                              aggfunc='sum').sort_values(by=sales_column, 
                                                         ascending=True)

    # барплот для поиска самых популярных жанров в регионе
    ax = df_genre.plot(kind='barh', legend=False, figsize=(19, 8), color ='green')

    for i in ax.patches:
        ax.text(i.get_width() + 0.2, i.get_y() + i.get_height() / 2,
                str(round((i.get_width()), 3)),
                fontsize=14,
                color='black')
    plt.title(f'{region}: Количество продаж игр по жанрам за 2012-2016 гг.'+ "\n", fontsize=20)
    plt.xlabel('Количество проданных копий игр, млн', fontsize=15)
    plt.ylabel('Жанры', fontsize=15)
    plt.show()

    # сводная таблица для определения влияния рейтинга ESRB на продажи в регионе
    df_rating = df.query('rating != "unk"').pivot_table(index='rating', 
                                                        values=sales_column, 
                                                        aggfunc='sum').sort_values(by=sales_column, 
                                                                                   ascending=True)
    
    # барплот для определения влияния рейтинга ESRB на продажи в регионе
    ax = df_rating.plot(kind='barh', legend=False, figsize=(19, 8), color ='orange')

    for i in ax.patches:
        ax.text(i.get_width() + 0.2, i.get_y() + i.get_height() / 2,
                str(round((i.get_width()), 3)),
                fontsize=14,
                color='black')
    plt.title(f'{region}: Влияние рейтинга ESRB на продажи'+ "\n", fontsize=20)
    plt.xlabel('Количество проданных копий игр, млн', fontsize=15)
    plt.ylabel('Категории рейтинга ESRB', fontsize=15)
    plt.show()
```

### Проверка гипотез

#### Гипотеза 1: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

***Сформулируем нулевую и альтернативную гипотезы:***

H0 (нулевая гипотеза): Средние пользовательские рейтинги платформ Xbox One и PC равны.

H1 (альтернативная гипотеза): Средние пользовательские рейтинги платформ Xbox One и PC не равны.

В данном случае нам нужно проверить гипотезу о равенстве средних двух генеральных совокупностей. Будем использовать t-тест Стьюдента. 

```python
# напишем функцию для проверки гипотезы о равенстве средних двух генеральных совокупностей

def mean_hypothesis(ho, h1, alpha, sample_1, sample_2, equal_var=True):
    """
    Функция проверяет гипотезу о равенстве средних, используя t-критерий Стьюдента.
    
    Parameters
    ----------
    ho: string
        Нулевая гипотеза.
    h1: string
        Альтернативная гипотеза.
    alpha: float
        Уровень значимости.        
    sample_1: Pandas Dataframe
        Набор данных 1.
    sample_2: Pandas Dataframe
        Набор данных 1.
    equal_var: string
        True - значение по умолчанию. Параметр метода ttest_ind.
        
    """
    
    # выводим гипотезы, уровень значимости
    print(f'H0 (нулевая гипотеза): {ho}')
    print(f'H1 (альтернативная гипотеза): {h1}')
    print(f'Уровень значимости: {alpha}')
    print('')
    
    # выводим среднее с дисперсию по сэмплам
    print('Sample_1')
    print(f'Среднее значение: {sample_1.mean():.2f}, дисперсия: {sample_1.var():.2f}')
    print('')
    
    print('Sample_2')
    print(f'Среднее значение: {sample_2.mean():.2f}, дисперсия: {sample_2.var():.2f}')
    print('')
    
    # применяем t-критерий Стьюдента 
    results = st.ttest_ind(sample_1, sample_2, equal_var=equal_var)
    print(f'p-значение: {results.pvalue:.2f}')
    print('')
    
    if results.pvalue < alpha:
        print('Отвергаем нулевую гипотезу')
    else:
        print('Не получилось отвергнуть нулевую гипотезу')
```

```python
# создадим датафреймы с данными платформ
xbox_one = actual_data.query('(platform == "XOne")&(user_score != -1)')
pc = actual_data.query('(platform == "PC")&(user_score != -1)')
```

```python
# проверим гипотезу №1
mean_hypothesis("Средние пользовательские рейтинги платформ Xbox One и PC равны.", 
                "Средние пользовательские рейтинги платформ Xbox One и PC не равны.", 
                0.05, 
                xbox_one['user_score'], 
                pc['user_score'], 
                equal_var=True)
```

***Вывод:***

Средние значения в выборках очень близки, в результате t-теста не удалось отвергнуть нулевую гипотезу, а значит есть основания полагать, что средние пользовательские рейтинги платформ Xbox One и PC равны.
